```python
# Statistical test
import scipy.stats as stats

category = combined.select_dtypes(include=object)

signif_cats = []
for i in category:
    tb1 = pd.crosstab(train.loc[:,i],train.is_claim)
    teststats,pvalue,df,exp_freq =  stats.chi2_contingency(tb1)
    #print("Pvalue for ",i,"is",pvalue)
    if(pvalue<0.05):
        print(i)
        signif_cats.append(i)
```

```python
num_cols = combined.select_dtypes(include=np.number).columns
```

```python
len(num_cols)
```

```python
num_cols
```

```python
nrows =4
ncols =3
iterator =1
for i in num_cols:
    plt.subplot(nrows,ncols,iterator)
    sns.boxplot(combined.is_claim,combined.loc[:,i])
    plt.title(i)
    iterator = iterator+1
    
plt.tight_layout()
plt.show()
```

```python
# statistical test on num vs cat(target)

signif_num=[]

for i in num_cols:
    zero = train.loc[train.is_claim==0,i]
    one = train.loc[train.is_claim==1,i]
    teststats,pvalue = stats.ttest_ind(zero,one)
    if pvalue<0.05:
        signif_num.append(i)
        print(i)
```

```python
# consolidate all the features
final_data = combined.loc[:,signif_cats + signif_num]
```

```python
final_data.columns
```

```python
final_data.shape
```

```python
# Feature Engineering......
final_data.head()
```

```python
# yes-1,no-0
for col in final_data.columns:
    if col.startswith('is'):
        final_data[col].replace({"Yes":1,"No":0},inplace=True)
```

```python
final_data.head()
```

```python
# max torque and max power

final_data[['Nm','torque']]=final_data.max_torque.str.split('@',expand=True)
```

```python
# max torque and max power

final_data[['bhp','power']]=final_data.max_power.str.split('@',expand=True)
```

```python
final_data.head()
```

```python
final_data['Nm']=final_data.Nm.apply(lambda x: float(x[:-2]))
final_data['torque'] = final_data.torque.apply(lambda x: float(x[:-3]))
```

```python
final_data['bhp']= final_data.bhp.apply(lambda x: float(x[:-3]))
final_data['power'] = final_data.power.apply(lambda x: float(x[:-3]))
```

```python
final_data.head()
```

```python
final_data.head(1)

# width,Nm,torque,bhp,power,population_density
```

```python
# speed~RPM =(30*Power)/(3.14*Torque)
final_data['speed']=final_data.torque/final_data.Nm
```

```python
final_data.speed
```

```python
zero = final_data.speed.loc[final_data.speed==0]
one = final_data.speed.loc[final_data.speed==1]
teststats,pvalue = stats.ttest_ind(zero,one)
print(teststats,pvalue)
```

```python
final_data.speed.loc[final_data.speed==0]
```

```python
list_engines=final_data.engine_type.value_counts()[0:3].index
```

```python
def engine_types(x):
    if x in list_engines:
        return('Top 3')
    else:
        return('Others')
```

```python
final_data['engine_types']=final_data.engine_type.apply(engine_types)
```

```python
final_data.drop(['max_torque','max_power','engine_type'],axis=1,inplace=True)
```

```python
final_data.head(1)
```

```python
#split the data in train and test

newtrain =  final_data.loc[0:train.shape[0]-1,:]
newtest =  final_data.loc[0:test.shape[0]:,:]
```

```python
train.shape,test.shape,newtrain.shape,newtest.shape
```

```python
# drop the target from newtest
newtest.drop('is_claim',axis=1,inplace=True)
```

```python
!pip install category-encoders
```

```python
import category_encoders as ce

tgt_encoding = ce.TargetEncoder(cols=['area_cluster','segment','model'])
X = newtrain.drop('is_claim',axis=1)
y = newtrain.is_claim

newx = tgt_encoding.fit_transform(X,y)
```

```python
newtest = tgt_encoding.transform(newtest)
```

```python
newx = pd.get_dummies(newx,drop_first=True)
newtest = pd.get_dummies(newtest,drop_first=True)
```

```python
newx.head()
```

```python
# StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
list_cols =['population_density', 'width', 'Nm', 'torque',
       'bhp', 'power', 'speed']

#sc.fit_transform(pd.DataFrame())

for i in list_cols:
    newx.loc[:,i] = sc.fit_transform(pd.DataFrame(newx.loc[:,i]))
    newtest.loc[:,i] = sc.fit_transform(pd.DataFrame(newtest.loc[:,i]))
    
```

```python
newx.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from xgboost import XGBClassifier


# initiate the machines

logreg = LogisticRegression()
dtree = DecisionTreeClassifier()
rf = RandomForestClassifier()
gbm = GradientBoostingClassifier()
xgb = XGBClassifier()

# initiate the KFold
kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=0)
```

```python
# Tuning the Data as per Smote

from imblearn.over_sampling import SMOTE

smote = SMOTE()
newx,y=smote.fit_resample(newx,y)
```

```python
# KFold Cross Validation
pred=[]
for train_index,test_index in kfold.split(newx,y):
    xtrain = newx.loc[train_index]
    ytrain = y.loc[train_index]
    pred.append(xgb.fit(xtrain,ytrain).predict(newtest))
```

```python
# Submission file
finalpred = pd.DataFrame(pred).T.mode(axis=1)
submissionfile = pd.DataFrame({'policy_id':test.policy_id,'is_claim':0})

submissionfile['is_claim']=finalpred
```

```python
submissionfile.to_csv('XGBoost_smote.csv',index=False)
```

```python
from catboost import CatBoostClassifier

cboost=CatBoostClassifier()

pred=[]

for train_index,test_index in kfold.split(newx,y):
    xtrain=newx.loc[train_index]
    ytrain=y.loc[train_index]
    pred.append(cboost.fit(xtrain,ytrain).predict(newtest))
    
finalpred=pd.DataFrame(pred).T.mode(axis=1)
submission=pd.DataFrame({'policy_id':test['policy_id'],'is_claim':0})

submission['is_claim']=finalpred

submission.to_csv('CBOOST.csv',index=False)
```

```python
# stacking Model

# Catboost Parameter Tuning

from sklearn.model_selection import GridSearchCV

cboost = CatBoostClassifier()
params=[{'learning_rate':[0.01,0.1,0.001],'depth':[2,4,6,8]}]
grid = GridSearchCV(estimator =cboost,param_grid=params,cv=5)
```

```python
# fit the model and find the best parameters

grid.fit(newx,y)
print(grid.best_params_)
```

```python
# Stacking models

basemodels = [("log",logreg),("gbm",gbm),('xgboost',xgb)]
from sklearn.ensemble import StackingClassifier

stack = StackingClassifier(estimators = basemodels,final_estimator=cboost,cv=3)

pred=[]

for train_index,test_index in kfold.split(newx,y):
    xtrain=newx.loc[train_index]
    ytrain=y.loc[train_index]
    pred.append(stack.fit(xtrain,ytrain).predict(newtest))
    
finalpred=pd.DataFrame(pred).T.mode(axis=1)
submission=pd.DataFrame({'policy_id':test['policy_id'],'is_claim':0})

submission['is_claim']=finalpred

submission.to_csv('Stacking.csv',index=False)
```

```python
from lightgbm import LGBMClassifier

lgbm = LGBMClassifier(learning_rate=0.01)

pred=[]

for train_index,test_index in kfold.split(newx,y):
    xtrain=newx.loc[train_index]
    ytrain=y.loc[train_index]
    pred.append(lgbm.fit(xtrain,ytrain).predict(newtest))
    
finalpred=pd.DataFrame(pred).T.mode(axis=1)
submission=pd.DataFrame({'policy_id':test['policy_id'],'is_claim':0})

submission['is_claim']=finalpred

submission.to_csv('LGBM.csv',index=False)
```
